```python
#Q-2:

Handling missing data is crucial in data preprocessing,
as it can significantly impact the quality and reliability
of  analysis. Here are several common techniques 
for handling missing data and their advantages and disadvantages:

1. Complete Case Analysis (CCA):
   - Advantage: It is simple and straightforward to implement. 
               If dataset has only a few missing values, you can simply remove the
                rows with missing data and work with the complete cases.
   - Disadvantage: It can lead to a loss of valuable information, 
    especially if the missing data are not missing completely at random (MCAR). 
    Removing rows with missing data may introduce bias and reduce the 
    representativeness of the remaining data.

2. Mean/Median/Mode Imputation:
   - Advantage: It is a simple technique where missing values are replaced with the mean, 
                 median, or mode of the respective feature. This method preserves the overall distribution 
                 and can be applied to both numerical and categorical data.
   - Disadvantage: It underestimates the variability in the data and does 
    not account for any relationships or dependencies between variables. 
    This technique can distort statistical measures and potentially affect downstream analysis.

3. Hot-Deck Imputation:
   - Advantage: Hot-Deck imputation fills missing values by borrowing from similar 
observations within the dataset. This approach maintains the overall patterns and 
relationships in the data. It can be particularly useful when there is a clear 
grouping or clustering of similar records.
   - Disadvantage: It assumes that similar cases have similar values, which may 
    not always hold true. If the dataset does not exhibit clear patterns or groups, 
    the imputation may introduce biases. Additionally, it can be computationally 
    expensive for large datasets.

4. Multiple Imputation:
   - Advantage: Multiple Imputation generates several plausible imputations
for each missing value, taking into account the uncertainty associated with 
missing data. It captures the variability and preserves the relationships
between variables, leading to more accurate and reliable results.
   - Disadvantage: Multiple imputation can be computationally intensive, 
    especially for large datasets. Additionally, it requires assumptions 
    about the missing data mechanism (e.g., Missing At Random, Missing Not 
    At Random) and may require more advanced statistical techniques.

5. Model-Based Imputation:
   - Advantage: Model-based imputation builds a predictive model to 
estimate missing values based on the observed data. It utilizes the
relationships between variables and can lead to more accurate imputations.
   - Disadvantage: Model-based imputation relies on the assumption that the
    model accurately captures the underlying data generation process. 
    If the model is misspecified or there are strong dependencies not 
    captured by the model, the imputations may be biased. 
    It can also be computationally expensive, especially for complex models.

It's important to choose an appropriate imputation technique based on the
nature and characteristics of your data, the amount of missing data,
and the goals of your analysis. Each technique has its own advantages
and disadvantages, and there is no one-size-fits-all solution. 
Careful consideration and evaluation of the imputation method's
impact on your specific analysis are necessary to ensure reliable results.
```
